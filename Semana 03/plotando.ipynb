# Vamos Plotar alguns gráficos usando o Plotly

```python
import pandas as pd
from pathlib import Path
import plotly.express as px
pd.options.plotting.backend = "plotly"


arq = Path(".").parent / "arquivo_otimizado.csv"
df = pd.read_csv(arq)
df
```

> Gráfico de valor de salário desejado

```python
local = df.plot(kind="hist", x="job_location")
local.show()
```

> Gráfico da distribuição de trabalhadores por região

```python
reg = df["pay_period"].value_counts().reset_index()
reg.columns = ["Loc","Quant"]
reg["Loc"] = reg["Loc"].map({1:"HOURLY",2:"MONTHLY", 3:"YEARLY"})
reg = px.pie(data_frame=reg, names="Loc", values="Quant", title="Porcentagem de trabalhador por região")
reg.show()
```

> Gráfico de pessoas que recebem anúncio

```python
anunc = df["sponsored"].value_counts().reset_index()
anunc.columns = ["Aceitou", "Respostas"]
anunc = px.pie(data_frame=anunc, title="Porcentagem de pessoas com anúncio", names="Aceitou", values="Respostas")
anunc.show()
```

> Gráfico do nível de cada um dos candidatos

```python
lev = df["level"].value_counts().reset_index()
lev.columns = ["lev","quant"]
lev = lev.plot(kind="barh",y="lev",x="quant")
lev.show()
```
