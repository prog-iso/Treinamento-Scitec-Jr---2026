```python
print("""Vamos criar uma lista.\n
        \tPara adicionar um elemento, digite um numero e enter logo depois.
            \tQuando quiser parar de adicionar elementos, escreva \"pare\" e aperte enter\n\n""")

num = input("Digite um numero: ");

lista = []

while(num != "pare"):
    lista += [int(num)]
    num = input("Digite um numero: ")

alertas = []
print("Os numeros digitados foram:", end="")
for x in lista:
    print(f"| {x} |", end="")
    if x > 100: alertas += [x]

print("\n\nOs valores maiores que 100 da lista acima são: ")
if alertas == []: print("Não tem nenhum maior que 100 !!!")
else:
    for elem in alertas: print(elem)

if alertas == []: print("Sistema Seguro")
else: print("Revisar valores")
```
