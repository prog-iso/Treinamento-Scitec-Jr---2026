-> Definindo caminho do arquivo e criando o dataframe

```python
from pathlib import Path
import pandas as pd

arq = Path(".").parent / "LinkedIn_RDB_three.csv"

data = pd.read_csv(arq)
data
```

-> Verificando como está distribuida a coluna "pay_period"

```python
data["pay_period"].value_counts()
```

-> Trocando valores em str por valores em int8, para reduzir o tamanho do arquivo do dataframe

```python
data["pay_period"] = data["pay_period"].map({"HOURLY":1,"MONTHLY":2, "YEARLY":3}).astype("int8")
data["pay_period"].dtype
```

```python
data.info(memory_usage="deep")
```

-> Verificando a possibilidade de mexer na coluna "work_type"

```python
data["work_type"].value_counts()
```

```python
data["work_type"] = data["work_type"].map({"Other":0, "Internship":1, "Temporary":2, "Part-time":3, "Contract":4, "Full-time":5}).astype("int8")
data["work_type"].dtype
```

```python
data.info(memory_usage="deep")
```

-> Verificando se dá para trabalhar com a coluna "compensation"

```python
data["compensation"].value_counts()
```

-> Transformando as strings em números

```python
data["compensation"] = data["compensation"].str.replace("$","").str.replace(",","")
data["compensation"] = pd.to_numeric(data["compensation"], errors="coerce").astype("Float32")
data["compensation"].dtype
```

```python
data["compensation"].describe()
```

-> Reduzindo os valores, sem perder muita informação (ou mantendo o que mais importa), e comprimindo um pouco mais

```python
data["compensation"] = data["compensation"]/100
data["compensation"] = data["compensation"].astype("float16")
data["compensation"].describe()
```

```python
data.info(memory_usage="deep")
```

-> Verifincando e já trabalhando com a coluna "views"

```python
data["views"].describe()
```

```python
data["views"] = data["views"].astype("int16")
data.info(memory_usage="deep")
```

-> Verificando se dá para trabalhar com a coluna "ben_pack_id"

```python
data["ben_pack_id"].describe()
```

-> Verificando se só tem valores inteiros, apesar das variaveis serem do tipo float

```python
for x in data["ben_pack_id"]:
    if x%1 != 0: print(x)
```

```python
data["ben_pack_id"] = data["ben_pack_id"].astype("UInt8")
data.info(memory_usage="deep")
```

-> Verificando se dá para trabalhar no company_id e otimizando essa parte dos dados

```python
data["company_id"].describe()
```

```python
data["company_id"] = data["company_id"].astype("UInt32")
data.info(memory_usage="deep")
```

-> Verificando e trabalhando na coluna "level"

```python
data["level"].value_counts()
```

```python
data["level"] = data["level"].map({"Internship":0, "Executive":1, "Director":2, "Associate":3, "Entry level":4, "Mid-Senior level":5}).astype("Int8")
data.info(memory_usage="deep")
```

```python
data["job_domain"].describe()
```

-> Mostrando como os elementos serão associados por números e, depois, substituindo os elementos por números

```python
coluna = data["job_domain"].astype("category").cat.categories
coluna = {key:elem for key,elem in enumerate(coluna)}
print(coluna)
```

```python
data["job_domain"] = data["job_domain"].astype("category").cat.codes.astype("int8")
data.info(memory_usage="deep")
```

-> Verificamos que cada elemento de job_id é único e que podemos otimizar essa parte o dataframe

```python
data["job_id"].nunique()
```

```python
data["job_id"] = data["job_id"].astype("uint16")
data.info(memory_usage="deep")
```

-> Considerando que a descrição do que cada um faz,bem como o titulo deles na empresa, não é tão relevante para a analise dos dados do dataframe, vou retirar essas partes. Ainda mais que, caso seja necessário, com o job_id podemos recuperar quem é cada indivíduo

```python
data.pop("title")
data.pop("description")
data.info(memory_usage="deep")
```
